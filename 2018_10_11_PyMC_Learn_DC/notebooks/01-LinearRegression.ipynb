Now, let's import the `LinearRegression` model from the `pymc-learn` package.

```python
import pmlearn
from pmlearn.linear_model import LinearRegression
print('Running on pymc-learn v{}'.format(pmlearn.__version__))
```

Generate Synthetic Data
===

```python
X = np.random.randn(1000, 1)
noise = 2 * np.random.randn(1000, 1)
Y = 4 * X + 3 + noise
```

```python
Y = np.squeeze(Y)
```

```python
plt.scatter(X, Y)
```

Fit w/ ADVI
===

```python
LR = LinearRegression()
```

```python
LR.fit(X, Y, minibatch_size=100)
```

```python
LR.plot_elbo()
```

```python
Y_predict = LR.predict(X)
```

```python
LR.score(X, Y)
```

```python
traceplot(LR.trace)
```

```python
LR.summary
```

```python
max_x = max(X)
min_x = min(X)

m = LR.summary['mean']['betas__0_0']
b = LR.summary['mean']['alpha__0']
fig1 = plt.figure()
#ax = fig.add_subplot(111)
plt.scatter(X, Y)
plt.plot([min_x, max_x], [m*min_x + b, m*max_x + b], 'r', label='ADVI')
plt.legend()
```

Fit w/ NUTS
===

```python
LR2 = LinearRegression()
```

```python
LR2.fit(X, Y, inference_type='nuts', inference_args={'draws': 2000})
```

```python
LR2.score(X, Y)
```

```python
traceplot(LR2.trace)
```

Compare the two methods
===

```python
max_x = max(X)
min_x = min(X)

m = LR.summary['mean']['betas__0_0']
b = LR.summary['mean']['alpha__0']

m2 = LR2.summary['mean']['betas__0_0']
b2 = LR2.summary['mean']['alpha__0']
fig1 = plt.figure()
plt.scatter(X, Y)
plt.plot([min_x, max_x], [m*min_x + b, m*max_x + b], 'r', label='ADVI')
plt.plot([min_x, max_x], [m2*min_x + b2, m2*max_x + b2], 'g', label='NUTS', alpha=0.5)
plt.legend()
```
